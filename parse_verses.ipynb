```python
import pandas
```

```python
import xml.etree.ElementTree as ET
```

```python
def find_words(x):
    for child in x:
        if child.tag == 'w':
            yield child
        else:
            for word in find_words(child):
                yield word
```

```python
nouns = []
finite_verbs = []
infinite_verbs = []
for book_file in ['01_matthew_full', '02_mark_full', '03_luke_full', '04_john_full']:
    print(book_file)
    book = ET.parse('GNT_annotation_v1.0/%s.xml' % (book_file,))
    book_root = book.getroot()
    for word in find_words(book_root):
        pos = word.attrib['pos']
        print(word.attrib['{http://www.w3.org/XML/1998/namespace}id'], word.attrib['lemma'])
        if pos == 'noun': 
            nouns.append({'id': word.attrib['{http://www.w3.org/XML/1998/namespace}id'], 
                          'lemma': word.attrib['lemma'], 
                          'gender': word.attrib['gender'],
                          'case': word.attrib['case'],
                          'number': word.attrib['number'],
                          'domains': word.attrib['domains'].split(),
                          'subdomains': word.attrib.get('subdomains', '').split()
                         })
        elif pos == 'finite':
            finite_verbs.append({'id': word.attrib['{http://www.w3.org/XML/1998/namespace}id'], 
                          'lemma': word.attrib['lemma'], 
                          'voice': word.attrib['voice'],
                          'mood': word.attrib['mood'],
                          'number': word.attrib['number'],
                          'tense': word.attrib['tense'],
                          'person': word.attrib['person'],
                          'domains': word.attrib.get('domains', '').split(),
                          'subdomains': word.attrib.get('subdomains', '').split()})
        elif pos == 'infinitive':
            infinite_verbs.append({'id': word.attrib['{http://www.w3.org/XML/1998/namespace}id'], 
                          'lemma': word.attrib['lemma'], 
                          'voice': word.attrib['voice'],
                          'tense': word.attrib['tense'],
                          'domains': word.attrib['domains'].split(),
                          'subdomains': word.attrib['subdomains'].split()})
```

```python
def extract_bookname(s): return s.split('.')[1]
def extract_chapter(s): return s.split('.')[2]
def extract_verse(s): return s.split('.')[3]
def extract_verseref(s): 
    na28, book, chapter, verse, wordnum = s.split('.')
    return "%s %s:%s" % (book, chapter, verse)
```

```python
all_nouns_df = pandas.DataFrame.from_records(nouns)
all_nouns_df['book'] = all_nouns_df.id.map(extract_bookname)
all_nouns_df['chapter'] = all_nouns_df.id.map(extract_chapter)
all_nouns_df['verse'] = all_nouns_df.id.map(extract_verse)
all_nouns_df['verseref'] = all_nouns_df.id.map(extract_verseref)
```

# Koine Greek Word-Embeddings

There are some words with a domain but not subdomain.

For example:
- σιαγών (cheek, 8)
- μίλιον (million, 81)
- νηστεία (fasting, 53)

```python
subdomain_less = all_nouns_df[all_nouns_df.subdomains.map(lambda x: x == [])][['lemma', 'domains']]
subdomain_less['domain'] = subdomain_less.domains.map(lambda x: x[0] + ".0")
subdomain_less['lemmas'] = subdomain_less.lemma.map(lambda x: [x])
all_subdomains = subdomain_less.set_index('domain').lemmas.to_dict()
for lemma, sds in zip(all_nouns_df.lemma, all_nouns_df.subdomains):
    for subdomain in sds:
        if subdomain not in all_subdomains:
            all_subdomains[subdomain] = [lemma]
        elif lemma not in all_subdomains[subdomain]:
            all_subdomains[subdomain].append(lemma)
            
            
all_subdomains
```

```python
word_path = pandas.DataFrame({'subdomain_path': list(all_subdomains.keys()),
                 'words': list(all_subdomains.values())
                 })
word_path['domain'] = word_path.subdomain_path.str.split('.').map(lambda x: x[0]).map(int)
word_path['subdomain'] = word_path.subdomain_path.str.split('.').map(lambda x: x[1]).map(int)
word_path
```
